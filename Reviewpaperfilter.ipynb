```python
import pandas as pd

# Load the CSV file with a different encoding
try:
    df = pd.read_csv('/content/attempt 3.csv', encoding='utf-8')  # Try UTF-8 first
except UnicodeDecodeError:
    try:
        df = pd.read_csv('/content/attempt 3.csv', encoding='latin1')  # Fallback to Latin1
    except UnicodeDecodeError:
        df = pd.read_csv('/content/attempt 3.csv', encoding='windows-1252')  # Fallback to Windows-1252

# Define INCLUDE keywords (short-term and peak load forecasting)
include_keywords = [
    'short-term load forecasting', 'peak load forecasting', 'electric load peak',
    'heating demand peak', 'cooling demand peak', 'time of peak', 'peak demand prediction',
    'load curve analysis', 'daily peak load', 'hourly load forecasting', 'real-time load forecasting',
    'intraday load forecasting', 'load profile prediction', 'peak shaving', 'demand response',
    'load forecasting for peak periods', 'peak load management', 'load forecasting for grids',
    'energy demand spikes', 'peak electricity demand', 'peak load prediction',
    'load forecasting for utilities', 'peak load analysis', 'forecasting peak demand',
    'short-term energy forecasting'
]

# Filter rows where the abstract or title contains INCLUDE keywords
include_filter = df['Abstract'].str.contains('|'.join(include_keywords), case=False, na=False) | \
                 df['Article Title'].str.contains('|'.join(include_keywords), case=False, na=False)

# Apply the filter
filtered_df = df[include_filter]

# Save the filtered results to a new CSV file
filtered_df.to_csv('filtered_literature.csv', index=False)

print(f"Filtered {len(filtered_df)} relevant studies.")
```

```python
# Save the excluded results to a separate CSV file
# Get excluded papers (rows that do not match the inclusion criteria)
excluded_df = df[~include_filter]
excluded_df.to_csv('excluded_literature.csv', index=False)
```
